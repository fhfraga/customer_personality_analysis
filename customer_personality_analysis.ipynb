A **Análise de Personalidade do Cliente** é uma análise detalhada dos clientes ideais para uma empresa. Ajuda uma empresa a entender melhor seus clientes e facilita a modificação de produtos de acordo com as necessidades, comportamentos e preocupações específicas de diferentes tipos de clientes.

A análise da personalidade do cliente ajuda uma empresa a modificar seu produto com base em seus clientes-alvo de diferentes tipos de segmentos de clientes. Por exemplo, em vez de gastar dinheiro para comercializar um novo produto para cada cliente no banco de dados da empresa, uma empresa pode analisar qual segmento de cliente tem maior probabilidade de comprar o produto e, em seguida, comercializar o produto apenas nesse segmento específico.

## 3.1 CRIANDO HIPÓTESES

Criar hipóteses possíveis é uma forma de tentar validar as informações que podem ser retirado dos dados. Baseado nas hipóteses criadas, podemos selecionar as hipóteses que são possíveis e valida-lás e criar uma lista final.

### HIPÓTESES
**1.** Clientes mais velhos devem comprar mais na loja

**2.** Clientes com maior grau de educação devem ter renda maior

**3.** Clientes casados devem comprar mais

**4.** Clientes com maiores rendas devem comprar mais

**5.** Cliente com reclamações devem comprar menos

**6.** Cliente devem comprar mais na promoção

**7.** Cliente mais jovens devem comprar mais na web 

```python
# valor gasto total
df3['mnt_total'] = (df3['MntFishProducts'] + df3['MntFruits'] 
                   + df3['MntGoldProds'] + df3['MntMeatProducts'] 
                   + df3['MntSweetProducts'] + df3['MntWines'])

# número de compras total
df3['num_total_purchases'] = (df3['NumCatalogPurchases'] 
                            + df3['NumStorePurchases'] 
                            + df3['NumWebPurchases'])

# data final + 50
df3['final_date'] = df3['Dt_Customer'].max() + pd.DateOffset(days=50)

# dias desde que a pessoa virou cliente
df3['total_days'] = (df3['final_date'] - df3['Dt_Customer']).dt.days

# período total que a pessoa é cliente da loja
df3['on_store'] = df3['total_days'] - df3['Recency']

# média de quantos dias a pessoa fez uma compra
df3['frequency'] = np.round(df3['on_store'] / df3['num_total_purchases'])

# remover cliente que não fizeram compras
df3 = df3[df3['num_total_purchases'] != 0]

# dia que o cliente se cadastrou
df3['day_customer'] = df3['Dt_Customer'].dt.day

# mês que o cliente se cadastrou
df3['month_customer'] = df3['Dt_Customer'].dt.month

# ano que o cliente se cadastrou
df3['year_customer'] = df3['Dt_Customer'].dt.year

# idade do cliente
df3['customer_age'] = df3['final_date'].dt.year - df3['Year_Birth']
```

# 6 REDUÇÃO DE DIMENSIONALIDADE

A redução de dimensionalidade refere-se a técnicas para reduzir o número de variáveis de entrada no conjunto de dados, mantendo o máximo possível da variação do conjunto de dados original. 

Reduzindo a dimensionalidade teremos um modelo com menor tempo para o treinamento, que gasta menos recursos computacionais, reduz problemas de overfitting, entre outros fatores.

```python
df6 = df5.copy()

# excluindo variáveis que não necessitamos
df6 = df6.drop(['ID', 'Dt_Customer', 'final_date', 'total_days', 'on_store',
                'age_group', 'income_group'], axis=1)
X = df6.copy()
```

## 6.1 PCA

`PCA` é uma técnica de redução de dimensionalidade linear. Ele transforma um conjunto de variáveis correlacionadas `(p)` em um número menor `(k)` de variáveis não correlacionadas chamadas de **componentes principais**, mantendo o máximo possível da variação no conjunto de dados original.

```python
#n_components = número de colunas
pca = PCA(n_components=X.shape[1])
principal_components = pca.fit_transform(X)

# plot variáveis explicadas
features = range(pca.n_components_)

fig, ax = plt.subplots(figsize=(14, 7))
y = pca.explained_variance_ratio_ * 100
ax.bar(features, y, color='mediumblue')
plt.suptitle('O quanto as variáveis explicam a variabilidade dos dados?',
             color='dimgray')
ax.set_xlabel('Variáveis', fontsize=18)
ax.set_ylabel('Porcentagem de variação explicada', fontsize=18)
plt.show()

# componentes do pca
df_pca = pd.DataFrame(principal_components)
```

Uma única variável explica mais de 70% da variabilidade dos dados, vamos ver agora outro gráfico

```python
fig, ax = plt.subplots(figsize=(14, 7))
ax.plot((np.cumsum(pca.explained_variance_ratio_) * 100) , color='mediumblue')
plt.suptitle('Somatório das explicações das variáveis para a variabilidade dos dados',
             color='dimgray')
ax.set_xlabel('Variáveis', fontsize=18)
ax.set_ylabel('Porcentagem de variação explicada', fontsize=18)
plt.show()
```

```python
print(f'Variância explicada pelo primeiro componente principal:\
 {np.cumsum(pca.explained_variance_ratio_ * 100)[0]:.2f} %')

print(f'variância explicada pelos 2 primeiros componentes principais:\
 {np.cumsum(pca.explained_variance_ratio_ * 100)[1]:.2f} %')

print(f'variância explicada pelos 3 primeiros componentes principais:\
 {np.cumsum(pca.explained_variance_ratio_ * 100)[2]:.2f} %')

print(f'variância explicada pelos 5 primeiros componentes principais:\
 {np.cumsum(pca.explained_variance_ratio_ * 100)[4]:.2f} %')

print(f'variância explicada pelos 10 primeiros componentes principais:\
 {np.cumsum(pca.explained_variance_ratio_ * 100)[9]:.2f} %')
```

Vemos que com apenas 10 variáveis conseguimos explicar a variabilidade dos dados em 96.36%, sendo 26 variáveis para explicar apenas 3.64%. Só com a dimensionalidade conseguimos reduzir nosso conjunto de dados para uma variabilidade de 96,36% em um conjunto com 27.7% do tamanho original.

# 8 HYPERPARAMETER FINE TUNING

## 8.1 WITHIN-CLUSTER SUM OF SQUARE (WCSS)

```python
wcss = []

for i in clusters:
    # definindo o modelo
    kmeans = KMeans(init='random', n_clusters=i, n_init=10, 
                   max_iter=300, random_state=RANDOM_STATE)
    
    # treinando o modelo
    kmeans.fit(X)
    
    # validação 
    wcss_iter = kmeans.inertia_
    
    # colocanod na lista
    wcss.append(wcss_iter)
```

```python
fig, ax = plt.subplots(figsize=(14, 7))
plt.plot(clusters, wcss, linestyle='--', marker='o',
                color='mediumblue')
plt.suptitle(' O método do cotovelo(The Ebow Method) ',
             color='dimgray')
ax.set_xlabel('Grupos', fontsize=18)
ax.set_ylabel('WCSS', fontsize=18)
plt.show()
```

```python
Elbow_M = KElbowVisualizer(KMeans(), k=clusters, timing=False)
Elbow_M.fit(X)
Elbow_M.show();
```

## 8.2 SILHOUETTE SCORE

```python
Elbow_M = KElbowVisualizer(KMeans(), k=clusters, metric='silhouette', timing=False)
Elbow_M.fit(X)
Elbow_M.show();
```

O método do cotovelo nos indica que o número perfeito para a divisão em grupos é 4, já o método da silhueta nos diz que é cinco. Nesse caso, como fizemos uma método do cotovelo anteriormente e ele fica próximo de escolher por decisão o 3, escolheremos o 4 como número para os grupos.
